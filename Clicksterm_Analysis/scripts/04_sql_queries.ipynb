```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("clickstream.db")

orders_df = pd.read_csv(
    r"C:\Users\sande\Desktop\HITESH PROJECT\data\orders.csv"
)

orders_df.to_sql("orders", conn, if_exists="replace", index=False)

print("Orders table loaded successfully")
```

```python
pd.read_sql("SELECT * FROM web_events", conn)
```

```python
pd.read_sql("""
SELECT COUNT(DISTINCT session_id) AS total_sessions
FROM web_events
""", conn)
```

```python
pd.read_sql("""
SELECT 
COUNT(DISTINCT o.session_id) * 1.0 /
COUNT(DISTINCT e.session_id) AS conversion_rate
FROM web_events e
LEFT JOIN orders o 
ON e.session_id = o.session_id
""", conn)
```

```python
pd.read_sql("""
SELECT device, COUNT(DISTINCT session_id) AS sessions
FROM web_events
GROUP BY device
""", conn)
```
