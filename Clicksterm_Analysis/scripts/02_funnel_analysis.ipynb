```python
import pandas as pd

events = pd.read_csv("C:\\Users\\sande\\Desktop\\HITESH PROJECT\\data\\clean_web_events.csv")

funnel_steps = ["page_view", "add_to_cart", "checkout", "purchase"]

funnel = (
    .apply(set)
    .reset_index()
)

for step in funnel_steps:
    funnel[step] = funnel["event"].apply(lambda x: step in x)

funnel_count = funnel[funnel_steps].sum()
print(funnel_count)

conversion_rate = funnel_count["purchase"] / funnel_count["page_view"]
print("Conversion Rate:", round(conversion_rate * 100, 2), "%")
```
